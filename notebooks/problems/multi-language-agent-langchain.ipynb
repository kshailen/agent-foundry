## 4. Starter Code (has a bug — fix it!)

```python
from langchain_openai import ChatOpenAI
from langchain.agents import AgentExecutor, create_tool_calling_agent
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.tools import tool

llm = ChatOpenAI(model="gpt-4o-mini")

@tool
def search_products(query: str) -> str:
    """Search for products in the catalog. Query must be in English."""
    products = {
        "laptop": "MacBook Pro 14\" - $1,999",
        "headphones": "Sony WH-1000XM5 - $349",
        "keyboard": "Keychron K2 - $89",
    }
    for key, val in products.items():
        if key in query.lower():
            return val
    return f"No products found for: {query}"

tools = [search_products]

# BUG: No language handling — agent mixes languages in tool calls and responses
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a shopping assistant. Help users find products."),
    ("human", "{input}"),
    ("placeholder", "{agent_scratchpad}"),
])

agent = create_tool_calling_agent(llm, tools, prompt)
executor = AgentExecutor(agent=agent, tools=tools)

# Test 1: Spanish user — should respond in Spanish, tool call in English
result1 = executor.invoke({"input": "Busco unos auriculares buenos"})
print("Spanish query:", result1["output"])

# Test 2: French user — should respond in French, tool call in English
result2 = executor.invoke({"input": "Je cherche un clavier mécanique"})
print("French query:", result2["output"])

# Test 3: English user — everything in English
result3 = executor.invoke({"input": "I need a good laptop"})
print("English query:", result3["output"])
```
