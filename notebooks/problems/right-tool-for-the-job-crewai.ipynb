## 4. Starter Code (has a bug — fix it!)

The code below has an intentional issue. Read the problem description and fix it.

```python
from crewai import Agent, Task, Crew
from crewai import LLM
from crewai.tools import tool

@tool
def weather_lookup(location: str) -> str:
    """Look up information."""  # BUG: Vague description
    return f"Weather in {location}: 72°F, sunny"

@tool
def web_search(query: str) -> str:
    """Search for things."""  # BUG: Vague description
    return f"Results for {query}: ..."

@tool
def calculator(expression: str) -> str:
    """Compute stuff."""  # BUG: Vague description
    return str(eval(expression))

@tool
def calendar_check(date: str) -> str:
    """Check things."""  # BUG: Vague description
    return f"Events on {date}: Team standup at 10am"

llm = LLM(model="gpt-4o-mini")

agent = Agent(
    role="Helpful Assistant",
    goal="Help users with various queries",
    backstory="You are a helpful assistant.",
    tools=[weather_lookup, web_search, calculator, calendar_check],
    llm=llm,
)

task = Task(
    description="Answer the following queries: 1) What's the weather in San Francisco? 2) What is 15% of 230? 3) Do I have any meetings on Friday?",
    expected_output="Accurate answers using the correct tools for each query",
    agent=agent,
)

crew = Crew(agents=[agent], tasks=[task])
result = crew.kickoff()
print(result)
```
