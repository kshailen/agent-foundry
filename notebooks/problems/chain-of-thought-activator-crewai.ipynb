```python
from crewai import Agent, Task, Crew
from crewai import LLM

llm = LLM(model="gpt-4o-mini")

# BUG: This instruction encourages direct answers without showing reasoning steps
agent = Agent(
    role="Math Assistant",
    goal="Solve math problems",
    backstory="You are a math assistant. Give the answer to the user's math problem. Be concise — just state the final number.",
    llm=llm,
)

task = Task(
    description="A store sells notebooks for $4 each. Sarah buys 3 notebooks and pays with a $20 bill. She then buys 2 pens at $1.50 each. How much money does she have left?",
    expected_output="The answer to the math problem",
    agent=agent,
)

crew = Crew(agents=[agent], tasks=[task])

result = crew.kickoff()
print(result)
```
