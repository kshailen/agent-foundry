## 4. Starter Code (has a bug — fix it!)

The code below has an intentional issue. Read the problem description and fix it.

```python
from agents import Agent, Runner
from agents.tool import function_tool
import time

@function_tool
def get_weather(city: str) -> str:
    """Get current weather for a city."""
    time.sleep(1)
    weather_data = {
        "new york": "72°F, Sunny",
        "london": "58°F, Cloudy",
        "tokyo": "68°F, Partly Cloudy",
    }
    return weather_data.get(city.lower(), f"Weather data not available for {city}")

# BUG: The agent calls get_weather sequentially for each city
agent = Agent(
    name="Weather Reporter",
    instructions="You are a weather assistant. Get the weather for each city requested.",
    tools=[get_weather],
)

start = time.time()
result = Runner.run_sync(agent, "What's the weather in New York, London, and Tokyo?")
print(f"Time: {time.time() - start:.1f}s")
print(result.final_output)
```
