```python
from crewai import Agent, Task, Crew

researcher = Agent(
    role="Research Specialist",
    goal="Research the given topic thoroughly",
    backstory="You are an expert researcher with deep knowledge.",
    verbose=True,
)

analyst = Agent(
    role="Data Analyst",
    goal="Analyze the research findings",
    backstory="You are a data analyst who finds insights in research.",
    verbose=True,
)

visualizer = Agent(
    role="Visualization Expert",
    goal="Create visual representations of the analysis",
    backstory="You turn data into compelling visuals.",
    verbose=True,
)

reporter = Agent(
    role="Report Writer",
    goal="Compile everything into a final report",
    backstory="You write clear, comprehensive reports.",
    verbose=True,
)

research_task = Task(
    description="Research AI trends for 2025",
    agent=researcher,
    expected_output="Research findings",
)

# BUG: If any task fails, the entire crew crashes with no partial results
analysis_task = Task(
    description="Analyze the research (NOTE: this will fail due to a simulated error)",
    agent=analyst,
    expected_output="Analysis results",
)

viz_task = Task(
    description="Create visualizations from the analysis",
    agent=visualizer,
    expected_output="Visualization descriptions",
)

report_task = Task(
    description="Write the final report combining all outputs",
    agent=reporter,
    expected_output="Final report",
)

crew = Crew(
    agents=[researcher, analyst, visualizer, reporter],
    tasks=[research_task, analysis_task, viz_task, report_task],
    verbose=True,
)

# Test: One agent failure crashes the entire crew
result = crew.kickoff()
print(result)
```
