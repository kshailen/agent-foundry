## 4. Starter Code (has a bug — fix it!)

The code below has an intentional issue. Read the problem description and fix it.

```python
from crewai import Agent, Task, Crew, Process
from crewai.tools import tool

@tool
def get_stock_price(ticker: str) -> str:
    """Get the current stock price for a ticker symbol."""
    prices = {"AAPL": "$185.23", "GOOGL": "$142.56", "MSFT": "$378.91"}
    return prices.get(ticker, f"No data for {ticker}")

# BUG: No topic enforcement — the financial advisor answers any question
financial_agent = Agent(
    role="Financial Advisor",
    goal="Help users with investment and finance questions.",
    backstory="You are a financial advisor that helps users with financial decisions.",
    tools=[get_stock_price],
)

# Test: The agent happily answers a medical question
task = Task(
    description="What are the symptoms of diabetes?",
    expected_output="A response to the user's question.",
    agent=financial_agent,
)

crew = Crew(agents=[financial_agent], tasks=[task], process=Process.sequential)
result = crew.kickoff()
print(result)
```
