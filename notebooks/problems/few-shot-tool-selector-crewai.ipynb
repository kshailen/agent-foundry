## 4. Starter Code (has a bug — fix it!)

The code below has an intentional issue. Read the problem description and fix it.

```python
from crewai import Agent, Task, Crew
from crewai.tools import tool

@tool
def get_stock_price(ticker: str) -> str:
    """Get the current stock price for a given ticker symbol."""
    prices = {"AAPL": "$178.50", "GOOGL": "$141.20", "MSFT": "$378.90"}
    return prices.get(ticker.upper(), f"No data for {ticker}")

@tool
def get_weather(city: str) -> str:
    """Get the current weather for a given city."""
    weather = {"new york": "72°F, Sunny", "london": "58°F, Cloudy", "tokyo": "68°F, Clear"}
    return weather.get(city.lower(), f"No weather data for {city}")

# BUG: No few-shot examples — agent can't decide when to use tools vs. answer directly
assistant = Agent(
    role="Assistant",
    goal="Answer the user's question",
    backstory="You are a helpful assistant with access to tools. Answer the user's question.",
    tools=[get_stock_price, get_weather],
)

# Test 1: Should use tool
task1 = Task(
    description="What is Apple's stock price right now?",
    expected_output="The current stock price",
    agent=assistant,
)
crew1 = Crew(agents=[assistant], tasks=[task1])
print("Tool query:", crew1.kickoff())

# Test 2: Should NOT use tool
task2 = Task(
    description="What is the capital of Japan?",
    expected_output="The answer to the question",
    agent=assistant,
)
crew2 = Crew(agents=[assistant], tasks=[task2])
print("Knowledge query:", crew2.kickoff())
```
