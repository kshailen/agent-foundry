## 4. Starter Code (has a bug — fix it!)

The code below has an intentional issue. Read the problem description and fix it.

```python
from agents import Agent, Runner
from agents.tool import function_tool

@function_tool
def search_products(query: str) -> str:
    """Search for products in the catalog. Query must be in English."""
    products = {
        "laptop": "MacBook Pro 14\" - $1,999",
        "headphones": "Sony WH-1000XM5 - $349",
        "keyboard": "Keychron K2 - $89",
    }
    for key, val in products.items():
        if key in query.lower():
            return val
    return f"No products found for: {query}"

# BUG: No language handling — agent mixes languages in tool calls and responses
agent = Agent(
    name="Shopping Assistant",
    instructions="You are a shopping assistant. Help users find products.",
    tools=[search_products],
)

# Test 1: Spanish user — should respond in Spanish, tool call in English
result1 = Runner.run_sync(agent, "Busco unos auriculares buenos")
print("Spanish query:", result1.final_output)

# Test 2: French user — should respond in French, tool call in English
result2 = Runner.run_sync(agent, "Je cherche un clavier mécanique")
print("French query:", result2.final_output)

# Test 3: English user — everything in English
result3 = Runner.run_sync(agent, "I need a good laptop")
print("English query:", result3.final_output)
```
