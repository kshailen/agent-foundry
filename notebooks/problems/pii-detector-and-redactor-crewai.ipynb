## 4. Starter Code (has a bug — fix it!)

The code below has an intentional issue. Read the problem description and fix it.

```python
from crewai import Agent, Task, Crew
from crewai.tools import tool

@tool
def lookup_customer(name: str) -> str:
    """Look up customer details by name."""
    return (
        f"Customer: {name}\n"
        f"Email: alice@example.com\n"
        f"Phone: 555-867-5309\n"
        f"Status: Active"
    )

# BUG: The agent returns raw PII (email, phone) with no redaction
lookup_agent = Agent(
    role="Customer Lookup",
    goal="Look up customer details and return them",
    backstory="You are a customer lookup assistant. Return the customer details you find.",
    tools=[lookup_customer],
)

task = Task(
    description="Find the details for Alice",
    expected_output="Customer details for Alice",
    agent=lookup_agent,
)

crew = Crew(agents=[lookup_agent], tasks=[task])
result = crew.kickoff()
print(result)
```
