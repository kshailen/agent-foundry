```python
from crewai import Agent, Task, Crew, Process
from crewai import LLM

llm = LLM(model="gpt-4o-mini")

# BUG: Only the final result is shown — user waits with no feedback
# TODO: Stream intermediate results so the user sees progress after each step
outliner = Agent(
    role="Outliner",
    goal="Create a research outline",
    backstory="Create a brief research outline. List 3-4 key areas to investigate.",
    llm=llm,
)

researcher = Agent(
    role="Researcher",
    goal="Write detailed research summaries",
    backstory="Write a detailed research summary for each area in the outline.",
    llm=llm,
)

reporter = Agent(
    role="Reporter",
    goal="Write a polished final report",
    backstory="Turn research into a polished final report with an introduction and conclusion.",
    llm=llm,
)

outline_task = Task(
    description="Create a research outline for: The impact of AI on healthcare",
    expected_output="A list of 3-4 key areas to investigate",
    agent=outliner,
)

research_task = Task(
    description="Write detailed research for each area",
    expected_output="Detailed research summary for each area",
    agent=researcher,
    context=[outline_task],
)

report_task = Task(
    description="Write a polished final report",
    expected_output="A polished report with introduction and conclusion",
    agent=reporter,
    context=[research_task],
)

# BUG: User only sees the final report after waiting for all 3 steps
crew = Crew(
    agents=[outliner, researcher, reporter],
    tasks=[outline_task, research_task, report_task],
    process=Process.sequential,
)

result = crew.kickoff()
print(result)
```
