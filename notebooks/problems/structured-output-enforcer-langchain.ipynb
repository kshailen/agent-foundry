## 4. Starter Code (has a bug — fix it!)

The code below has an intentional issue. Read the problem description and fix it.

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
import json

llm = ChatOpenAI(model="gpt-4o-mini")

# BUG: This prompt does not enforce structured JSON output
prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful question-answering assistant. Answer the user's question."),
    ("human", "{input}"),
])

chain = prompt | llm

# Test: The output should be a JSON object with keys: answer, confidence, sources
result = chain.invoke({"input": "What is the capital of France?"})
print(result.content)

# Verify: try parsing as JSON
try:
    parsed = json.loads(result.content)
    assert "answer" in parsed
    assert "confidence" in parsed
    assert "sources" in parsed
    print("Valid structured output!")
except (json.JSONDecodeError, AssertionError) as e:
    print(f"FAILED: Output is not valid structured JSON - {e}")
```
