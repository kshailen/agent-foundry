```python
from crewai import Agent, Task, Crew
from crewai.tools import tool

@tool
def search_web(query: str) -> str:
    """Search the web for information."""
    return f"Search results for: {query}"

# BUG: This agent allows hallucination of URLs
assistant = Agent(
    role="Research Assistant",
    goal="Answer questions thoroughly",
    backstory="You are a helpful research assistant.",
    tools=[search_web],
)

# Test: Ask something that might trigger hallucinated URLs
task = Task(
    description="What are the best resources to learn about AI agents?",
    expected_output="A helpful answer with resources",
    agent=assistant,
)

crew = Crew(agents=[assistant], tasks=[task])
result = crew.kickoff()
print(result)
```
