## 4. Starter Code (has a bug — fix it!)

The code below has an intentional issue. Read the problem description and fix it.

```python
from crewai import Agent, Task, Crew

# BUG: No entity extraction — the agent doesn't track named entities
assistant = Agent(
    role="Assistant",
    goal="Help users with their questions",
    backstory="You are a helpful assistant.",
)

history = []

def chat(user_input: str) -> str:
    history.append(f"User: {user_input}")
    task = Task(
        description="Conversation so far:\n" + "\n".join(history) + f"\n\nRespond to the latest message: {user_input}",
        expected_output="A helpful response",
        agent=assistant,
    )
    crew = Crew(agents=[assistant], tasks=[task])
    result = str(crew.kickoff())
    history.append(f"Assistant: {result}")
    return result

# Turn 1: Mentions people and places
print(chat("I met with Sarah and David in Tokyo last March to discuss the Q1 report."))

# Turn 2: References an entity from turn 1
print(chat("What city was my meeting in?"))

# Turn 3: More entities
print(chat("Next week I\'m flying to Berlin to meet Professor Mueller on January 15th."))

# Turn 4: Query about people
print(chat("List all the people I\'ve mentioned so far."))
```
