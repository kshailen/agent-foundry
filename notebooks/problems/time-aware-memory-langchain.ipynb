```python
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage, SystemMessage

llm = ChatOpenAI(model="gpt-4o-mini")

# BUG: Facts stored without timestamps — "yesterday" is meaningless
memory = []

def store_fact(fact: str):
    memory.append(fact)

def chat(user_input: str) -> str:
    store_fact(f"User said: {user_input}")
    facts_block = "\n".join(memory)
    messages = [
        SystemMessage(content=f"You are a helpful assistant. Known facts:\n{facts_block}"),
        HumanMessage(content=user_input),
    ]
    response = llm.invoke(messages)
    return response.content

# Monday
print(chat("I had a meeting with the design team today."))

# Tuesday
print(chat("I submitted the Q3 report today."))

# Wednesday — "yesterday" should mean Tuesday, not Monday
print(chat("What did I do yesterday?"))

# Query by range
print(chat("What happened in the last 3 days?"))

print(f"\nStored memory: {memory}")
```
