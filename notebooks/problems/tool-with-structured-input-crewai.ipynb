## 4. Starter Code (has a bug — fix it!)

The code below has an intentional issue. Read the problem description and fix it.

```python
from crewai import Agent, Task, Crew, Process
from crewai.tools import tool

# BUG: The tool takes a flat string for the address, which the agent formats inconsistently.
# Sometimes it sends "123 Main St, Springfield, IL 62701", other times "Springfield IL 123 Main".
# The downstream API expects structured fields and parsing a flat string is error-prone.
# TODO: Replace the flat string with structured input (street, city, state, zip)

@tool
def validate_address(address: str) -> str:
    """Validate a shipping address."""
    parts = address.split(",")
    if len(parts) < 3:
        return f"Invalid address format: '{address}'. Expected: street, city, state zip"
    return f"Address validated: {address}"

@tool
def calculate_shipping(address: str, weight_lbs: float) -> str:
    """Calculate shipping cost to an address."""
    return f"Shipping {weight_lbs}lbs to '{address}': $12.99 (parsed address may be wrong)"

shipping_agent = Agent(
    role="Shipping Assistant",
    goal="Help users validate addresses and calculate shipping costs.",
    backstory="You are a shipping assistant that helps with address validation and shipping calculations.",
    tools=[validate_address, calculate_shipping],
)

task = Task(
    description="Validate and get shipping cost for a 5lb package to 742 Evergreen Terrace, Springfield, IL 62704",
    expected_output="Address validation result and shipping cost estimate.",
    agent=shipping_agent,
)

crew = Crew(agents=[shipping_agent], tasks=[task], process=Process.sequential)
result = crew.kickoff()
print(result)
```
