```python
from langchain_openai import ChatOpenAI
from langgraph.graph import StateGraph, START, END
from typing import TypedDict

llm = ChatOpenAI(model="gpt-4o-mini")

class State(TypedDict):
    topic: str
    raw_data: str
    cleaned_data: str
    analysis: str
    report: str

def collect(state: State) -> dict:
    response = llm.invoke(f"Collect key data points about: {state['topic']}")
    return {"raw_data": response.content}

def clean(state: State) -> dict:
    response = llm.invoke(f"Clean and normalize this data: {state['raw_data']}")
    return {"cleaned_data": response.content}

def analyze(state: State) -> dict:
    response = llm.invoke(f"Analyze trends and insights from: {state['cleaned_data']}")
    return {"analysis": response.content}

def report(state: State) -> dict:
    response = llm.invoke(f"Generate an executive report from: {state['analysis']}")
    return {"report": response.content}

# BUG: The pipeline produces a wrong report but there's no way to tell which node erred
# TODO: Add tracing/logging to each node to identify the faulty one
graph = StateGraph(State)
graph.add_node("collect", collect)
graph.add_node("clean", clean)
graph.add_node("analyze", analyze)
graph.add_node("report", report)
graph.add_edge(START, "collect")
graph.add_edge("collect", "clean")
graph.add_edge("clean", "analyze")
graph.add_edge("analyze", "report")
graph.add_edge("report", END)

app = graph.compile()
result = app.invoke({"topic": "global renewable energy adoption trends", "raw_data": "", "cleaned_data": "", "analysis": "", "report": ""})
print(result["report"])
```
