## 4. Starter Code (has a bug — fix it!)

The code below has an intentional issue. Read the problem description and fix it.

```python
from crewai import Agent, Task, Crew
import time

# Simulated document corpus
documents = [f"Document {i}: This is a report about topic {chr(65 + i % 26)} covering various aspects of the subject including analysis, findings, and recommendations for Q{(i % 4) + 1} 2024." for i in range(20)]

analyst = Agent(
    role="Document Analyst",
    goal="Summarize and synthesize documents",
    backstory="You are a senior research analyst who distills large document sets into actionable reports.",
)

# BUG: Documents are processed sequentially — extremely slow for 20+ docs
# TODO: Implement map-reduce pattern with parallel document processing
def process_documents(docs: list[str]) -> str:
    start = time.time()
    tasks = []

    for doc in docs:
        task = Task(
            description=f"Summarize this document in 2-3 sentences: '{doc}'",
            expected_output="A 2-3 sentence summary",
            agent=analyst,
        )
        tasks.append(task)

    tasks.append(Task(
        description="Synthesize all the document summaries above into a cohesive executive report. Identify key themes and patterns.",
        expected_output="An executive summary report",
        agent=analyst,
    ))

    crew = Crew(agents=[analyst], tasks=tasks)
    result = crew.kickoff()

    elapsed = time.time() - start
    return f"[Processed {len(docs)} docs in {elapsed:.2f}s]\n{result}"

result = process_documents(documents)
print(result)
```
