## 4. Starter Code (has a bug — fix it!)

```python
from langchain_openai import ChatOpenAI
from langgraph.graph import StateGraph, START, END
from typing import TypedDict

llm = ChatOpenAI(model="gpt-4o-mini")

class State(TypedDict):
    topic: str
    research: str
    analysis: str
    recommendation: str

# BUG: All stages use the same framework
# TODO: Integrate stages from different frameworks
def research(state: State) -> dict:
    response = llm.invoke(f"Research: {state['topic']}")
    return {"research": response.content}

def process(state: State) -> dict:
    response = llm.invoke(f"Process this research: {state['research']}")
    return {"analysis": response.content}

def decide(state: State) -> dict:
    response = llm.invoke(f"Provide recommendations based on: {state['analysis']}")
    return {"recommendation": response.content}

graph = StateGraph(State)
graph.add_node("research", research)
graph.add_node("process", process)
graph.add_node("decide", decide)
graph.add_edge(START, "research")
graph.add_edge("research", "process")
graph.add_edge("process", "decide")
graph.add_edge("decide", END)

app = graph.compile()
result = app.invoke({"topic": "Best cloud provider for AI workloads", "research": "", "analysis": "", "recommendation": ""})
print(result["recommendation"])
```
