```python
from crewai import Agent, Task, Crew, Process
from crewai import LLM

llm = LLM(model="gpt-4o-mini")

# BUG: This backstory makes the agent guess instead of asking for clarification
assistant = Agent(
    role="Travel Assistant",
    goal="Help users book flights and hotels",
    backstory="You are a travel booking assistant. Help users book flights and hotels. Always provide a complete answer.",
    llm=llm,
)

queries = [
    "Book a flight",
    "Book a one-way flight from New York JFK to London Heathrow on March 15, 2025, economy class, for 1 adult",
    "I need a flight from San Francisco to Tokyo next week",
]
for q in queries:
    task = Task(description=q, expected_output="A booking response", agent=assistant)
    crew = Crew(agents=[assistant], tasks=[task], process=Process.sequential)
    result = crew.kickoff()
    print(f"Query: {q}\nResponse: {result}\n")
```
