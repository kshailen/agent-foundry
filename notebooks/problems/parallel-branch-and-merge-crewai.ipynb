```python
from crewai import Agent, Task, Crew
import time

# BUG: Sentiment and entity extraction run sequentially, doubling latency
# TODO: Run both analyses in parallel and merge results
analyst = Agent(
    role="Text Analyst",
    goal="Analyze text for both sentiment and entities",
    backstory="You are an NLP specialist.",
)

def analyze_text(text: str) -> dict:
    start = time.time()

    sentiment_task = Task(
        description=f"Analyze the sentiment of this text: '{text}'. Reply with positive, negative, or neutral and a brief explanation.",
        expected_output="Sentiment label and explanation",
        agent=analyst,
    )

    entity_task = Task(
        description=f"Extract all named entities (people, organizations, locations, products) from: '{text}'. Return as a JSON list.",
        expected_output="JSON list of entities",
        agent=analyst,
    )

    crew = Crew(agents=[analyst], tasks=[sentiment_task, entity_task])
    results = crew.kickoff()

    elapsed = time.time() - start
    return {"results": str(results), "processing_time": f"{elapsed:.2f}s"}

texts = [
    "Apple CEO Tim Cook announced the new iPhone at their Cupertino headquarters. Customers loved it.",
    "The project in Berlin failed miserably. Google and Microsoft both pulled out of the partnership.",
]
for text in texts:
    result = analyze_text(text)
    print(f"Text: {text[:50]}...\nResult: {result}\n")
```
