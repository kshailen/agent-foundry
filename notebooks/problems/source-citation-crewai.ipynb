## 4. Starter Code (has a bug — fix it!)

The code below has an intentional issue. Read the problem description and fix it.

```python
from crewai import Agent, Task, Crew
from crewai import LLM

llm = LLM(model="gpt-4o-mini")

documents = [
    {"content": "The company's revenue grew 40% year-over-year in Q3 2024.", "source": "earnings_report_q3.pdf"},
    {"content": "The new product launch is scheduled for March 2025.", "source": "product_roadmap.pdf"},
    {"content": "Employee satisfaction scores improved to 4.5/5 in the latest survey.", "source": "hr_annual_review.pdf"},
]

# BUG: The prompt does not ask the LLM to cite sources — answers have no attribution
context = "\n".join([doc["content"] for doc in documents])

def ask(question: str) -> str:
    agent = Agent(
        role="Research Assistant",
        goal="Answer questions based on provided documents",
        backstory=f"Answer the question based on the context below.\n\nContext: {context}",
        llm=llm,
    )
    task = Task(
        description=question,
        expected_output="An answer based on the provided context",
        agent=agent,
    )
    crew = Crew(agents=[agent], tasks=[task])
    return str(crew.kickoff())

print(ask("What was the revenue growth?"))
```
