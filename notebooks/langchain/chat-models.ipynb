## 3. Initialize a Chat Model

`init_chat_model` provides a single entry point for all supported providers.

```python
from langchain.chat_models import init_chat_model

model = init_chat_model("gpt-4o-mini", model_provider="openai")

response = model.invoke("What is LangChain?")
print(response.content)
```

## 4. Key Parameters

Control model behavior with `temperature`, `max_tokens`, and `timeout`.

```python
model = init_chat_model(
    "gpt-4o-mini",
    model_provider="openai",
    temperature=0.2,
    max_tokens=100,
    timeout=30,
)

response = model.invoke("Explain AI agents in one sentence.")
print(response.content)
print(f"\nTokens used: {response.usage_metadata}")
```

## 5. Invoke with a Message List

Pass a list of messages for multi-turn or system-prompted conversations.

```python
from langchain_core.messages import SystemMessage, HumanMessage

messages = [
    SystemMessage(content="You are a concise assistant. Reply in one sentence."),
    HumanMessage(content="What is a transformer model?"),
]

response = model.invoke(messages)
print(response.content)
```

## 7. Comparing Provider Responses

Use the same question across multiple models to compare outputs.

```python
question = "What is an AI agent? Answer in one sentence."

openai_model = init_chat_model("gpt-4o-mini", model_provider="openai")
print("OpenAI:", openai_model.invoke(question).content)
```
