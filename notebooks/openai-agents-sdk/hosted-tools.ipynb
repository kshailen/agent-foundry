## 3. WebSearchTool

Give your agent the ability to search the web for real-time information. Use `search_context_size` to control how much context is retrieved.

```python
from agents import Agent, Runner, WebSearchTool

agent = Agent(
    name="Research Agent",
    instructions="You are a research assistant. Search the web to answer questions with up-to-date information.",
    tools=[WebSearchTool()],
)

result = await Runner.run(agent, "What were the major AI announcements this week?")
print(result.final_output)
```

## 4. WebSearchTool with search_context_size

Control the amount of search context: `"low"` for speed, `"medium"` for balance, `"high"` for thoroughness.

```python
deep_agent = Agent(
    name="Deep Research Agent",
    instructions="You provide thorough, well-researched answers.",
    tools=[WebSearchTool(search_context_size="high")],
)

result = await Runner.run(deep_agent, "Explain the latest advances in quantum computing.")
print(result.final_output)
```

## 6. CodeInterpreterTool

Let your agent write and execute Python code in a sandboxed environment on OpenAI's servers.

```python
from agents import CodeInterpreterTool

analyst = Agent(
    name="Data Analyst",
    instructions="You analyze data and perform calculations. Use code to compute results.",
    tools=[CodeInterpreterTool()],
)

result = await Runner.run(analyst, "Calculate the first 20 Fibonacci numbers and find their sum.")
print(result.final_output)
```

## 7. ImageGenerationTool

Enable your agent to generate images from text descriptions.

```python
from agents import ImageGenerationTool

creative_agent = Agent(
    name="Creative Agent",
    instructions="You help users create images. Use the image generation tool.",
    tools=[ImageGenerationTool()],
)

result = await Runner.run(creative_agent, "Create an image of a sunset over a mountain lake.")
print(result.final_output)
```

## 8. Combining Multiple Hosted Tools

Agents can use several hosted tools together for powerful multi-capability workflows.

```python
multi_agent = Agent(
    name="Multi-Tool Agent",
    instructions=(
        "You are a research analyst. Search the web for data, "
        "then use code to analyze and visualize it."
    ),
    tools=[
        WebSearchTool(search_context_size="high"),
        CodeInterpreterTool(),
    ],
)

result = await Runner.run(
    multi_agent,
    "Find the current population of the 5 largest countries and create a comparison.",
)
print(result.final_output)
```

## 9. Fact-Checking Agent with Web Search

A practical example that uses web search to verify claims.

```python
fact_checker = Agent(
    name="Fact Checker",
    instructions=(
        "You are a fact-checking agent. When given a claim, search the web "
        "to verify it. Respond with VERIFIED, UNVERIFIED, or FALSE, "
        "followed by your evidence."
    ),
    tools=[WebSearchTool(search_context_size="high")],
)

result = await Runner.run(
    fact_checker,
    "Claim: Python is the most popular programming language in 2025.",
)
print(result.final_output)
```

## Key Takeaways

- Hosted tools run on OpenAI's infrastructure — no setup, servers, or API keys needed
- `WebSearchTool` gives agents real-time web access with configurable `search_context_size`
- `FileSearchTool` searches your uploaded documents via vector store IDs
- `CodeInterpreterTool` executes Python in a sandboxed environment for computation
- `ImageGenerationTool` creates images from text descriptions
