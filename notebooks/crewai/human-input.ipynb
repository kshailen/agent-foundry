## Task with `human_input=True` and crew `kickoff`

The copywriter agent drafts marketing copy; CrewAI pauses so you can review and send feedback before the task is finalized.

```python
from crewai import Agent, Crew, Process, Task

copywriter = Agent(
    role="Marketing Copywriter",
    goal="Write clear, persuasive copy that matches brand tone",
    backstory="You draft marketing copy, revise based on stakeholder feedback, and deliver polished final text.",
    verbose=True,
)

task = Task(
    description="Write a marketing copy for our new AI product",
    expected_output="A compelling 200-word marketing copy",
    agent=copywriter,
    human_input=True,
)

crew = Crew(
    agents=[copywriter],
    tasks=[task],
    process=Process.sequential,
    verbose=True,
)

result = crew.kickoff()
print(result.raw)
```

## Note on Colab and input prompts

In **Google Colab**, when the agent is ready for your review, Colab shows an **input prompt** in the cell output—type your feedback there and press Enter. In a plain terminal, the same prompt appears as **standard input** while the process waits.

## Key takeaways

- **`human_input=True`** on a **`Task`** pauses execution so a **human can review** and provide **feedback** before the final answer.
- The agent **incorporates** that feedback into the **final** task output.
- **Colab** surfaces this as an **input field** in the notebook; **CLI** runs use **stdin**.
- Use for **high-stakes**, **brand**, or **expert-gated** work rather than every task.
