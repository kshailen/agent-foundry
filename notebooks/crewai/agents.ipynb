## 3. Anatomy of an Agent

Every CrewAI agent has key properties: `role`, `goal`, `backstory`, `verbose`, and `allow_delegation`.

````python
from crewai import Agent

agent = Agent(
    role="Data Analyst",
    goal="Analyze datasets and extract actionable insights",
    backstory="You are a senior data analyst with 10 years of experience in business intelligence. You excel at finding patterns in data and communicating findings clearly.",
    verbose=True,
    allow_delegation=False,
)

print(f"Agent role: {agent.role}")
print(f"Agent goal: {agent.goal}")
````

## 4. Writing Effective Roles and Goals

Specific, professional roles and action-oriented goals produce dramatically better results.

````python
# A vague agent
vague_agent = Agent(
    role="Helper",
    goal="Help with code",
    backstory="You help people.",
    verbose=True,
)

# A specific agent — much better results
specific_agent = Agent(
    role="Senior Backend Engineer specializing in Python microservices",
    goal="Review Python code for bugs, security issues, and performance problems, providing specific fix suggestions",
    backstory="""You are a security-focused code reviewer with 8 years of experience
at top tech companies. You have deep expertise in OWASP Top 10 vulnerabilities
and Python-specific security pitfalls. You believe in actionable feedback —
every issue you flag comes with a concrete fix.""",
    verbose=True,
)

print(f"Vague agent role: {vague_agent.role}")
print(f"Specific agent role: {specific_agent.role}")
````

## 5. Agent with Tools

Agents become much more powerful when equipped with tools.

**Note:** `SerperDevTool` requires a [Serper API key](https://serper.dev/). Set `SERPER_API_KEY` below if you want to try it.

````python
from crewai import Task, Crew
from crewai_tools import SerperDevTool

# Uncomment and set if you have a Serper API key:
# os.environ["SERPER_API_KEY"] = getpass("Enter your Serper API key: ")

researcher = Agent(
    role="Market Researcher",
    goal="Gather competitive intelligence from the web",
    backstory="You are a market research analyst skilled at finding and synthesizing information from multiple online sources.",
    tools=[SerperDevTool()],
    verbose=True,
)

print(f"Researcher tools: {[t.__class__.__name__ for t in researcher.tools]}")
````

## 6. Run a Crew with the Specific Agent

Let's put the specific code reviewer agent to work on a simple task.

````python
review_task = Task(
    description="""Review the following Python function for bugs, security issues, and performance:

```python
def get_user(user_id):
    query = f"SELECT * FROM users WHERE id = {user_id}"
    result = db.execute(query)
    return result
```

Provide specific fixes for each issue found.""",
    expected_output="A list of issues found with severity, explanation, and code fix for each.",
    agent=specific_agent,
)

crew = Crew(
    agents=[specific_agent],
    tasks=[review_task],
    verbose=True,
)

result = crew.kickoff()
print(result)
````

## Key Takeaways

- Agents are defined by their **role**, **goal**, and **backstory**
- Specific, detailed descriptions produce dramatically better results
- Tools extend what an agent can do beyond just LLM reasoning
- `allow_delegation` controls whether agents can hand off work to teammates
